This study aims to provide a comprehensive diversity profile of the company, examining the role of race and representation across key aspects of employment. By analyzing factors such as salary, position, performance reviews, length of employment, and termination trends, this research seeks to identify potential relationships between race and workplace outcomes.  

Beyond understanding demographic composition, this analysis offers insight into whether implicit or explicit discrimination may exist, helping prospective employees make informed career decisions. The findings will contribute to a data-driven perspective on workplace equity, ensuring transparency in employment practices.  


```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Here I am looking to see if the company does employee non-US Citizens who are eligible to work in the US as well as those who are not 
#eligible to work in the US.
citizen_info = demographic_data.value_counts("CitizenDesc")
citizen_info
```

```python
citizen_percentages = (citizen_info / citizen_info.sum()) * 100
citizen_percentages
```

```python
citizen_percentages.plot(kind="pie", autopct="%1.1f%%", colors=["blue", "red", "green"], figsize=(6,6))
plt.title("Citizenship Breakdown")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
citizen_avg_salary = demographic_data.groupby("CitizenDesc")["Salary"].mean().sort_values(ascending=False)
print("Average Salary by Race")
citizen_avg_salary
```

```python
citizen_avg_salary.plot(kind="barh", color="skyblue", figsize=(8,5))

plt.title("Average Salary by Citizenship Status")
plt.xlabel("Average Salary ($)")
plt.ylabel("Citizenship Status")
plt.grid(axis="x", alpha=0.7)

plt.show()
```

As a US compnay, it is expected that the majority of employees will be US Citizens. However, the data shows that the company is 
willing to employee a little over 5% non US citizens; both who may or may not be eligible to work in the United States at the moment. 
This can be an indication that this company is willing to sponsor work visas or a path to obtaining work visas to those employees willing to work hard.

When we examine the bar graph to look at the salary comparison between the 3 groups, we can see non US citizens actually are paid more on average than US citizens. Which clearly indicates that there is no pay discrimintation occuring or employees are not being exploited due to their imigration status.

```python
#Here we will count the amount of employees by race.
race_df = demographic_data.value_counts("RaceDesc")
race_df
```

```python
race_counts = demographic_data["RaceDesc"].value_counts()
race_percentages = (race_counts / race_counts.sum()) * 100

# Loop through each race category and format output
for race, percentage in race_percentages.items():
    print(f"Percentage of {race}: {percentage:.2f}%")
```

```python
race_percentages.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6), colormap="Pastel1")
plt.title("Race Distribution in the Company")
plt.ylabel("")  # Hide y-label for clarity
plt.show()
```

```python
# Here we will calculate the average salary according to each race.
avg_sal_per_race = demographic_data.groupby("RaceDesc")["Salary"].mean().sort_values(ascending=False)
print("Average Salary by Race")
avg_sal_per_race
```

```python
avg_sal_per_race.plot(kind="barh", color="steelblue", figsize=(8,5), edgecolor="black")

plt.title("Average Salary by Race", fontsize=14, fontweight="bold")
plt.xlabel("Average Salary ($)", fontsize=12)
plt.ylabel("Race", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.7)

plt.show()
```

```python
#Here we will calculate the averarge performance score according to race.

avg_perf_score = demographic_data.groupby("RaceDesc")["PerfScoreID"].mean().sort_values(ascending=False)
print("Average Performance Score by Race")
avg_perf_score
```

```python
#Now we will visualize this data into a Bar Graph.
avg_perf_score.plot.barh()
plt.title("Average Performance Score per Race")
plt.xlabel("PerfScoreID")
plt.ylabel("Race")
```

```python
avg_perf_score.plot(kind="barh", color="mediumseagreen", figsize=(8,5), edgecolor="black")

# Formatting the chart
plt.title("Average Performance Score by Race", fontsize=14, fontweight="bold")
plt.xlabel("Performance Score", fontsize=12)
plt.ylabel("Race", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.7)

plt.show()
```
