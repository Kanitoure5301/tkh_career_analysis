```python
#Here I am looking to see if the company does employee non-US Citizens who are eligible to work in the US as well as those who are not 
#eligible to work in the US.
citizen_info = demographic_data.value_counts("CitizenDesc")
citizen_info
```

```python
#Here we will review how many employees identify as Hispanic/Lation or not.
hispanic_latino = demographic_data.value_counts("HispanicLatino")
hispanic_latino
```

In the code above, we see that since some responses in the data set begin with a lowercase 'y' and/or 'n', they are being interpreted as separate unique responses. We have to correct this in order to get an accurate percentage.

```python
demographic_data["HispanicLatino"] = demographic_data["HispanicLatino"].str.upper()
```

```python
hispanic_latino = demographic_data.value_counts("HispanicLatino")
hispanic_latino
```

Now we see that the count has been correctly adjusted by coverting all responses to all uppercase string characters.

```python
#Here we get the percentage of that count.
hispanic_counts = demographic_data["HispanicLatino"].value_counts()
hispanic_percentages = (hispanic_counts / hispanic_counts.sum()) * 100

for response, percentage in hispanic_percentages.items():
    print(f"Percentage of {response}: {percentage:.2f}%")
```

```python
#We can see this percentage visualized in the bar graph below.
plt.bar(hispanic_percentages.index, hispanic_percentages.values, color=['skyblue', 'salmon'])

plt.xlabel("Response")
plt.ylabel("Percentage")
plt.title("Hispanic/Latino Representation")

plt.show()
```

```python
#In this code block, we will find the average salary for those who identify as Hispanic/Latino compare to those who do not.
avg_latx_sal = demographic_data.groupby("HispanicLatino")["Salary"].mean()

print(avg_latx_sal)
```

```python

avg_latx_sal.plot(kind="bar", color=["skyblue", "salmon"])

plt.xlabel("Hispanic/Latino Identification")
plt.ylabel("Average Salary")
plt.title("Average Salary by Hispanic/Latino Identification")

plt.show()
```

```python
#Here we get the average performance review score for each group.
avg_latx_score = demographic_data.groupby("HispanicLatino")["PerfScoreID"].mean()

print(avg_latx_score)
```

```python
avg_latx_score.plot(kind="barh", color="mediumseagreen", figsize=(8,5))

plt.title("Average Performance Score by Hispanic/Lation Identification", fontsize=14, fontweight="bold")
plt.xlabel("Performance Score", fontsize=12)
plt.ylabel("Identifies as Hispanic/Latino", fontsize=12)

plt.show()
```

```python
#Here we will count the amount of employees by race.
race_df = demographic_data.value_counts("RaceDesc")
race_df
```

```python
#In this code block we will determine the percentage of each racial group within the company.
race_counts = demographic_data["RaceDesc"].value_counts()
race_percentages = (race_counts / race_counts.sum()) * 100

for race, percentage in race_percentages.items():
    print(f"Percentage of {race}: {percentage:.2f}%")
```

```python
#Here we can see this visualized in a pie chart.
race_percentages.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6), colormap="Pastel1")
plt.title("Race Distribution in the Company")
plt.ylabel("") 
plt.show()
```

```python
white_median_salary = demographic_data[demographic_data["RaceDesc"] == "White"]["Salary"].median()
minority_median_salary = demographic_data[demographic_data["RaceDesc"] != "White"]["Salary"].median()

print(white_median_salary)
print(minority_median_salary)
```

```python
equity_ratio = minority_median_salary / white_median_salary
print(f"Equity Ratio: {equity_ratio:.2f}")
```

```python
# Here we will calculate the average salary according to each race.
avg_sal_per_race = demographic_data.groupby("RaceDesc")["Salary"].mean().sort_values(ascending=False)
print("Average Salary by Race")
avg_sal_per_race
```

```python
avg_sal_per_race.plot(kind="barh", color="steelblue", figsize=(8,5))

plt.title("Average Salary by Race", fontsize=14, fontweight="bold")
plt.xlabel("Average Salary ($)", fontsize=12)
plt.ylabel("Race", fontsize=12)

plt.show()
```

```python
#Here we will calculate the averarge performance score according to race.

avg_perf_score = demographic_data.groupby("RaceDesc")["PerfScoreID"].mean().sort_values(ascending=False)
print("Average Performance Score by Race")
avg_perf_score
```

```python
avg_perf_score.plot(kind="barh", color="mediumseagreen", figsize=(8,5))

plt.title("Average Performance Score by Race", fontsize=14, fontweight="bold")
plt.xlabel("Performance Score", fontsize=12)
plt.ylabel("Race", fontsize=12)

plt.show()
```

```python
# Here we will get the count of termination reason across each specific racial group.
termination_percentages = (
    demographic_data.groupby("RaceDesc")["TermReason"]
    .value_counts(normalize=True) * 100
)

print(termination_percentages)
```

```python
# Here I have excluded the rows where "TermReason" is "N/A-StillEmployed". This will allow us to exclude those who are still employed with the company.
filtered_data = demographic_data[demographic_data["TermReason"] != "N/A-StillEmployed"]
filtered_data
```

```python
#Now we can see the percentage of each termination reason for each race.
termination_percentages = (
    filtered_data.groupby("RaceDesc")["TermReason"]
    .value_counts(normalize=True) * 100
)

print(termination_percentages)
```

```python
termination_df = termination_percentages.unstack()

termination_df.plot(kind="bar", stacked=True, figsize=(10,6), colormap="viridis")

plt.xlabel("Race")
plt.ylabel("Percentage of Terminations")
plt.title("Termination Reasons Across Racial Groups")
plt.legend(title="Termination Reason", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.show()
```

Question 3: Are performance assessments distributed fairly across racial demographics?

To answer this question, I analyzed performance review scores across racial groups. Based on the data, there is no observable relationship between race and performance evaluations. Across all racial groups, average review scores remain consistently between 2.7 and 3.3, indicating that race does not appear to be a determining factor in assessment outcomes.

Question 4: Are there any statistically significant differences in termination rates that could suggest bias or inequitable workplace conditions?

While employees of different races cited various reasons for leaving the company, the data does not indicate any termination patterns associated with racial bias orinequitable workplace conditions.

Question 5: Is there a statistically significant difference in average salary across racial groups?

Although White employees make up the majority of the workforce, the data does not indicate that they receive preferential compensation compared to other racial groups.
